```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
#基于轮廓系数选取最好的n_clusters
#图一知道每个聚出来的类的轮廓系数，还想要一个各个轮廓系数之间的对比
#图二知道聚类之后图像分布的模样
```

```python
n_clusters=4
clusterer = KMeans(n_clusters=n_clusters,random_state=10).fit(X)
clusterer_labels =clusterer.labels_
silhouette_avg=silhouette_score(X,clusterer_labels)
sample_silhouette_values = silhouette_samples(X, clusterer_labels)
print('For n_clusters:',n_clusters,'The average silhouette_score is:',silhouette_avg)
```

```python
y_lower = 10
#创建画布
fig,(ax1,ax2)=plt.subplots(1,2)
#画布尺寸
fig.set_size_inches(18,7) 
ax1.set_xlim([-0.1,1])
ax1.set_ylim([0,X.shape[0]+(n_clusters+1)*10])
for i in range(n_clusters):
    ith_cluster_silhouette_values = sample_silhouette_values[clusterer_labels==i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    
    ### cm的nipy_spectral函数根据色素板0-1之间的浮点数来生成颜色
    color = cm.nipy_spectral(float(i)/n_clusters)
    
    ### fill_between是一个让一个范围内的柱状图颜色都统一的函数
    ### fill_betweenx的范围是在纵坐标上，设置横向柱状图
    ### fill_betweeny的范围是在横坐标上
    ### fill_betweenx(纵坐标的上限，纵坐标的下线,x轴上的取值，柱状图的颜色) 、、、此处画图
    ax1.fill_betweenx(np.arange(y_lower, y_upper)
                      ,ith_cluster_silhouette_values
                      ,facecolor=color
                      ,alpha=0.7) #alpha=0.7透明度
    ax1.text(-0.05
             , y_lower + 0.5 * size_cluster_i
             , str(i))#（内容显示位置横坐标，纵坐标，内容）
    y_lower = y_upper + 10
ax1.set_title("The silhouette plot for the various clusters.")
ax1.set_xlabel("The silhouette coefficient values")
ax1.set_ylabel("Cluster label")
ax1.axvline(x=silhouette_avg, color="red", linestyle="--") #画平均值
ax1.set_yticks([])
ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
colors = cm.nipy_spectral(clusterer_labels.astype(float) / n_clusters)
ax2.scatter(X[:, 0], X[:, 1]
            ,marker='o'
            ,s=8
            ,c=colors)
centers = clusterer.cluster_centers_
# Draw white circles at cluster centers
ax2.scatter(centers[:, 0], centers[:, 1], marker='x',
            c="red", alpha=1, s=200)
ax2.set_title("The visualization of the clustered data.")
ax2.set_xlabel("Feature space for the 1st feature")
ax2.set_ylabel("Feature space for the 2nd feature")
#全图设置标题
plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
              "with n_clusters = %d" % n_clusters),
             fontsize=14, fontweight='bold')
plt.show()
```

### 确定初始质心
### random_state:一个数对应一种随机生成初始质心方式的种子
### init：默认‘k-means++’初始设置彼此远离的质心，得到结果更为可靠，最好不要动，'random'随机，还可以输入（n_cluster,n_feature）就是簇数，各个特征数，相当于直接给出质心
### n_init:整数，默认10，使用不同的质心随机初始化的种子来运行k-means算法的次数。最终结果会是基于Inertia来计算的n_init次连续运行后的最佳输出

### max_iter：整数，默认300，单次运行的k-means算法的最大迭代次数
### tol：浮点数，默认1e-4，两次迭代间Inertia下降的量，如果两次迭代之间Inertia下降的值小于tol所设定的值，迭代就会停下

```python
random = KMeans(n_clusters=10,init='random',max_iter=10,random_state=420).fit(X)
y_pred_max10 =random.labels_
silhouette_score(X,y_pred_max10)
```

```python
random = KMeans(n_clusters=10,init='random',max_iter=20,random_state=420).fit(X)
y_pred_max10 =random.labels_
silhouette_score(X,y_pred_max10) #竟然提升了轮廓系数的表现
```

### 实例：聚类算法用于降维，KMeans的矢量量化应用.
特征选择的降维是直接选取对模型贡献最大的特征，PCA的降维是聚合信息，而矢量量化的降维是在同等样本量上压缩信息的大小，即不改变特征的数目也不改变样本的数目，只改变在这些特征下的样本上的信息量。

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans 
from sklearn.datasets import load_sample_image #导入两张图片的库
from sklearn.utils import shuffle #打乱一个序列的函数，比如打乱一个列表，字典，多维数组等
from sklearn.metrics import pairwise_distances_argmin #对两个序列中的点进行距离匹配的库
```

```python
#KMeans把九万种颜色聚类为64类，还不太影响图片
n_clusters = 64
#归一化处理，plt.imshow 在浮点数上的表现非常优异，因此我们把china中的数据转化成浮点数
china = np.array(china,dtype=np.float)/china.max()
```

```python
(china<0).sum()
```

```python
(china<0).sum()
```

```python
w,h,d=original_shape=tuple(china.shape) #将china转换成矩阵格式
```

```python
assert d==3 #强制要求d为3，不然报错
```

```python
image_array = np.reshape(china,(w*h,d)) #改变形状
```

```python
#因为样本比较大先随机使用1000个数找出质心,再用找出的质心为质心
image_array_sample = shuffle (image_array,random_state=0)[:1000] 
kmeans = KMeans(n_clusters=n_clusters,random_state=0).fit(image_array_sample)
kmeans.cluster_centers_
```

```python
labels =kmeans.predict(image_array)
labels.shape
```

```python
#用质心来替换样本
images_kmeans = image_array.copy()
for i in range(w*h):
    images_kmeans[i] = kmeans.cluster_centers_[labels[i]]
```

```python
pd.DataFrame(images_kmeans).drop_duplicates().shape #只有64种颜色了
```

```python
#恢复图片结构
kmean_image = np.reshape(images_kmeans,(w,h,d))
```

```python
kmean_image.shape
```

```python
centroid_random = shuffle(image_array,random_state=0)[:n_clusters]
```

```python
centroid_random.shape
```

```python
labels_random = pairwise_distances_argmin(centroid_random,image_array,axis=0)
#函数pairwise_distances_argmin(x1,x2,axis)函数分别是序列
#用来计算x2中的每个样本到x1中的每个样本的距离，返回和x2形状相同，与x1所对应的最近样本的索引
```

```python
labels_random.shape
```

```python
len(set(labels_random))
image_random = image_array.copy()
for i in range(w*h):
    image_random[i] = centroid_random[labels_random[i]]
image_random = image_random.reshape(w,h,d)
image_random.shape
```

```python
plt.figure(figsize=(10,10))
plt.axis('off')#不显示坐标轴
plt.title('Original image (96,615 colors)')
plt.imshow(china) #原本的颜色
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('Quantized image (64 colors, K-Means)')
plt.imshow(kmean_image) #聚类的颜色
plt.figure(figsize=(10,10))
plt.axis('off')
plt.title('Quantized image (64 colors, Random)')
plt.imshow(image_random)
plt.show() #随机质心对应的图片
```
