### random_state&splitter

```python
clf=tree.DecisionTreeClassifier(criterion='entropy',random_state=30) #建立树模型 
#criterion='entropy'信息熵，criterion='gini'是基尼系数, random_state=30随机模式30sklearn经常使用不同的模式建立树，可以固定模式随便选则
clf=clf.fit(xtrain,ytrain) #进行训练
score=clf.score(xtest,ytest)#导入测试集打分,返回预测精准度
score

#random_state用来设置随机模式参数，默认None，在高纬度时随机性会表现更加明显，而低维度数据(如鸢尾花数据集)随机性几乎不会显现
#不显现则输入任意数会长出同一颗树而让模型稳定.
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy',splitter='random') #建立树模型 
#criterion='entropy'信息熵，criterion='gini'是基尼系数, random_state=30随机模式30sklearn经常使用不同的模式建立树，可以固定模式随便选则
clf=clf.fit(xtrain,ytrain) #进行训练
score=clf.score(xtest,ytest)#导入测试集打分,返回预测精准度
score
#splitter也是用来控制决策树的随机选项，有两种输入值，输入'best'，决策数在分支时虽然随机，但还是会选择更重要的特征进行分支（
#重要性可以透过（feature_importances_）查看，输入'random'，决策树在分支时会更加随机，树会更深，对训练集的拟合将会更低.这也是
#防止过拟合的一种方式.
```

```python
feature_names=wine.feature_names
dot_data = tree.export_graphviz(clf,feature_names=feature_names,class_names=['桃花酒','菊花酒','梅子酒'],filled=True,rounded=True,out_file = None)
graph=graphviz.Source(dot_data)
graph
```

## min_samples_leaf & min_samples_split

```python
#min_samples_leaf限定，一个节点在分支之后的每子个节点都必须包含min_samples_leaf个训练样本，不然此节点不继续分支，或者
#分支会照着每个字节点都包含min_simple_leaf个样本方案去发生。
#一般的搭配max_depth使用，有奇效，使得决策树更加平滑，其是一条树干最对进行几次分支的限定
#min_samples_split限定，一个节点必须包含min_samples_split个训练样本，不然此节点不继续分支.
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy'
                                ,random_state=30
                                ,splitter='random'
                                ,max_depth=3
                                ,min_samples_leaf=10
#                                 ,min_samples_split=10
                               )
clf=clf.fit(xtrain,ytrain) 
feature_names=wine.feature_names
dot_data = tree.export_graphviz(clf,feature_names=feature_names,class_names=['桃花酒','菊花酒','梅子酒'],filled=True,rounded=True,out_file = None)
graph=graphviz.Source(dot_data)
graph
```

```python
score=clf.score(xtrain,ytrain)
score1=clf.score(xtest,ytest)
print(score)
print(score1)
```

## max_features & min_impurity_decrease

```python
#max_features一般用于树“精修”，
#max_features限制分支时考虑的特征个数，超过限制个数的特征都会被舍弃.和max_depth有异曲同工之妙，
#max_feature是用来限制高纬度的过拟合的减支操作，但其方法暴力，直接强制限制，可能会导致决策树学习不足.如果希望通过降维的方法防止过拟合，建议使用PCI和ICA和特征选择模块的降维方法.
#min_impurity_decrease限制信息增益的大小，信息增益小于设定值(贡献值小于一定值)的分支不会发生.
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy'
                                ,random_state=30
                                ,splitter='random'
                                ,min_impurity_decrease=0.1
                               )
clf=clf.fit(xtrain,ytrain) 
feature_names=wine.feature_names
dot_data = tree.export_graphviz(clf,feature_names=feature_names,class_names=['桃花酒','菊花酒','梅子酒'],filled=True,rounded=True,out_file = None)
graph=graphviz.Source(dot_data)
graph
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy'
                                ,random_state=30
                                ,splitter='random'
                                ,max_features=7
                               )
clf=clf.fit(xtrain,ytrain) 
feature_names=wine.feature_names
dot_data = tree.export_graphviz(clf,feature_names=feature_names,class_names=['桃花酒','菊花酒','梅子酒'],filled=True,rounded=True,out_file = None)
graph=graphviz.Source(dot_data)
graph
```

```python
import matplotlib.pyplot as plt
test=[]
for i in range(10):
    clf=tree.DecisionTreeClassifier(criterion='entropy'
                                    ,random_state=30
                                    ,splitter='random'
                                    ,max_depth=i+1
                                    ,max_features=5
                                   )
    clf=clf.fit(xtrain,ytrain) 
    score=clf.score(xtest,ytest)
    test.append(score)
plt.plot(range(1,11),test,color='red',label="max_depth")
plt.legend() #标签
plt.show()
```
